```python
#importing libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings=
```

```python
df=pd.read_csv("employee_dataset_for_eda.csv")
df
```

```python
#employee count per department
sns.countplot(data=df,x='Department',palette='Set2')
plt.title("employee count per department")
```

finance department having greater than count of employee

```python
#gender distribution
sns.countplot(data=df,x="Gender",palette="coolwarm")
plt.title("gender distribution")
```

```python
#employee loccation distribution
sns.countplot(data=df,x="Location",palette="Spectral")
plt.title("Number of Employees by Location")
plt.xticks(rotation=45)
```

Employees work in seattle more than other

```python
#Age distribution
sns.histplot(df['Age'],bins=10,kde=True)
plt.title("Age distribution")
```

45-50 is greater than followed by 25-30

```python
#years of experience distribution
sns.histplot(df["Years_of_Experience"],bins=10,kde=True)
plt.title("Years_of_Experience")
```

year 2.5 - 5.0 is greater followed by 12.5



```python
#monthly salary distribution
sns.boxplot(data=df,y="Monthly_Salary",color="darkpink")
plt.title("salary spread")
```
